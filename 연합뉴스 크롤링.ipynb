```python
# 함수로 
# 크롤링한 제목, 링크 넣을 리스트 생성
headlines_list =[]
href_list = [] 

def yeonhap_page(input_page):
    

    url_base = 'https://www.yna.co.kr/economy/all/{}'.format(input_page)
    page = urlopen(url_base)
    yeonhap_soup = BeautifulSoup(page, 'html.parser')

        # 가져올 구역 (왼쪽 최근기사면) 추출
    yeonhap_soup = yeonhap_soup.find('section', class_='box-type box-latest01')



        # 링크 추출

    
    
    yeonhap_news = yeonhap_soup.find_all("strong", class_="tit-news")
    yeonhap_href = yeonhap_soup.find_all("a", class_='tit-wrap')
    
    
    
    headlines_list.extend([news.string for news in yeonhap_news])
    href_list.extend(['https:' + href['href']for href in yeonhap_href])

    data = {
              '제목': headlines_list
            , '링크': href_list
     }
    df = pd.DataFrame(data)
    return df
```

```python
for i in range(1,21): # 20페이지까지 존재
    all = yeonhap_page(i)

all
```
