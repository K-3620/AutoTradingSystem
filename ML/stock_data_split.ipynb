```
# import package

import pandas as pd
import numpy as np
import math
```

```
# read csv file -> daily stock data code 'A000020'

df = pd.read_csv('/content/drive/My Drive/datasets/stock/DA000020.csv')
```

```
# check Dataframe

df.head(10)
```

```
# copy DataFrame

df_stock = df.copy()
```

```
# transform 'date'column datatype to datetime -Function

def transform_date(df):
  df['date'] = df['date'].astype('str')
  df['date'] = pd.to_datetime(df['date'])
  return df 
```

```
# transform df_stock and check df_stock infomation

df_stock = transform_date(df_stock)
df_stock.info()
```

```
# set index 'date' and check index

df_stock.set_index(df_stock['date'],inplace=True)
df_stock.index
```

```
# drop data column and df_stock Sort by past
df_stock.drop('date',axis=1,inplace=True)
df_stock.sort_values(by='date',inplace=True)
```

```
# make ['label'] colomn

df_stock['label'] = df_stock['close'].shift(-1)
```

```
# copy df_stock [0:len(df_stock)] -> except last row
n = len(df_stock)
print(n)
df_data = df_stock.iloc[:n-1]
df_data.tail(10)
```

```
# create X_features => 1 column 'close' y_label => 1 column 'label'
X_features = df_data['close']
y_label = df_data['label']
```

```
X_features.tail(10)
```

```
y_label.tail(10)
```

```
# split train , test data
data_len = len(X_features)
train_len = math.ceil(data_len*0.8)
X_train = X_features[:train_len]
X_test = X_features[train_len:]
y_train = y_label[:train_len]
y_test = y_label[train_len:]
print('X_tratin shape : {}, X_test shape : {} , y_train : {}, y_test : {}'.format(X_train,X_test,y_train,y_test))
```

```
df_train = pd.DataFrame()
df_train['X_train'] = X_train
df_train['y_train'] = y_train

df_test = pd.DataFrame()
df_test['X_test'] = X_test
df_test['y_test'] = y_test
```

```
# save csv file if you want save csv file , you can use this code


df_train.to_csv('/content/drive/My Drive/datasets/stock/train.csv')
df_test.to_csv('/content/drive/My Drive/datasets/stock/test.csv')
```
